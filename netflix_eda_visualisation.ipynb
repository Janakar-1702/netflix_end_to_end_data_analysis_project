# **IMPORTING NECESSARY LIBRARIES**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **IMPORTING THE DATASET**

```python
df=pd.read_csv('/content/netflix_data.csv', encoding='latin1')
```

```python
df.head()
```

```python
df.describe()
```

# **1. Content Strategy Visualization**
**Movies vs TV Shows**

```python
sns.countplot(data=df, x='type')
plt.title("Movies vs TV Shows Distribution")
plt.show()
```

# **2. Netflix Growth Over Time**

```python
year_counts = df['year_added'].value_counts().sort_index()

year_counts.plot(kind='line')
plt.title("Content Added Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

# **3. Catalog Freshness**

```python
sns.countplot(data=df, x='content_category')
plt.title("Content Category Distribution")
plt.xticks(rotation=45)
plt.show()
```

# **4. Content Age Distribution**

```python
sns.histplot(df['content_age'], bins=30)
plt.title("Content Age Distribution")
plt.show()
```

# **5. Genre Investment Strategy**

```python
top_genres = df['Genre'].value_counts().head(10)

sns.barplot(x=top_genres.values, y=top_genres.index)
plt.title("Top 10 Genres on Netflix")
plt.show()
```

# **6. Global Content Strategy**

```python
top_country = df['country'].value_counts().head(10)

sns.barplot(x=top_country.values, y=top_country.index)
plt.title("Top Producing Countries")
plt.show()
```

# **7. Audience Targeting**

```python
sns.countplot(data=df, y='rating')
plt.title("Content Rating Distribution")
plt.show()
```
